```python
import random
import time

# Генерация массива заданного размера
def generate_array(size):
    return [random.randint(1, 1000) for _ in range(size)]

# Сортировка массива методом Шелла
def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2

# Размеры массивов для эксперимента
array_sizes = [100, 500, 1000, 5000]

# Сортировка массивов различных размеров и измерение времени выполнения
for size in array_sizes:
    array = generate_array(size)
    print(f"Исходный массив размером {size}: {array[:5]} ... {array[-5:]}")
    
    start_time = time.time()
    shell_sort(array)
    end_time = time.time()
    
    print(f"Отсортированный массив размером {size}: {array[:5]} ... {array[-5:]}")
    print(f"Время выполнения сортировки: {end_time - start_time:.6f} секунд\n")
```

```python
import random
import time

# Генерация массива заданного размера
def generate_array(size):
    return [random.randint(1, 1000) for _ in range(size)]

# Сортировка массива с помощью алгоритма сортировки пузырьком
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Размеры массивов для эксперимента
array_sizes = [100, 500, 1000, 5000]

# Измерение времени выполнения сортировки для каждого размера массива
for size in array_sizes:
    array = generate_array(size)
    
    start_time = time.time()
    sorted_array = bubble_sort(array)
    end_time = time.time()
    
    execution_time = end_time - start_time
    print(f"Сортировка массива размером {size} заняла {execution_time:.6f} секунд")
```

```python
import time
import matplotlib.pyplot as plt

# Recursive function to implement tournament algorithm
def tournament_algorithm(arr):
    if len(arr) == 1:
        return arr[0]
    left = tournament_algorithm(arr[:len(arr)//2])
    right = tournament_algorithm(arr[len(arr)//2:])
    if arr[0] > arr[len(arr)//2]:
        return left
    else:
        return right

# Generate data for different array sizes
sizes = []
times = []
for i in range(10, 10000, 100):
    arr = list(range(i))
    start = time.time()
    tournament_algorithm(arr)
    end = time.time()
    sizes.append(i)
    times.append(end - start)

# Plot the graph
plt.plot(sizes, times)
plt.xlabel('Array size (n)')
plt.ylabel('Time taken (s)')
plt.title('Time complexity of tournament algorithm')
plt.show()
```
